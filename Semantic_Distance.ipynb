```python
d = []
for each in data:
    # print each
    d.append(each.split(';'))
```

```python
words = []
modes = []
for each in d:
    #print len(each[3]) # number of Chars per 'story'
    if len(each[3]) > 0:
        words.append(each[3].split())
        modes.append(each[0])
```

```python
cleanedStories = []
i = 0
for each in words:
    len(each) # The number of words in a story as determined by split on spaces
    cleanedStories.append([])
    for word in each:
        foundStopWord = False
        for stopWord in giantListOfStopWords:
            if findWholeWord(stopWord)(word) != None:
                foundStopWord = True
                break
        if not foundStopWord:
            word = re.sub(r'\W+', '', word) #remove anything that is not a-z...
            word = word.lower()
            # print word
            if len( wn.synsets(word) ) == 0:
                pass
            else:
                cleanedStories[i].append(word)
    i += 1
            
i = 0
cleanedStories.pop(0)
cleanedStories.pop(0)
modes.pop(0)
modes.pop(0)
for each in cleanedStories:
    print i , 'length of story ', len(each)
    i += 1
```

```python
names = []
i = 0
j = 0
modesToKeep = []
for each in cleanedStories:
    if len(each) > 25:
        modesToKeep.append( modes[j] )
        names.append( [] )
        for word in each:
            names[i].append( wn.synsets(word)[0].name)
        i += 1
    j += 1
```

```python
k = 0
avgSemDisAll = 0
simCount = 0
novelty = []
for story in names:
    # print story
    storyLength = len(story)
    #print 'The length of the story is: ', storyLength
    j = 0
    i = 0
    avgSemDistStory = 0
    storySimCount = 0
    for each in story:
        i = j+1
        #print 'Path Similarity ', each
        while i < storyLength:
            sim =  wn.path_similarity(  wn.synset(each) , wn.synset(story[i])  )
            if sim != None:
                simCount += 1
                storySimCount += 1
                avgSemDisAll = avgSemDisAll + sim
                avgSemDistStory = avgSemDistStory + sim
            i += 1
        j += 1
    novelty.append( (avgSemDistStory,storySimCount) )
    k += 1
```
